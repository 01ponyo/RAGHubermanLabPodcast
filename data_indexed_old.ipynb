```python
read_me = "README.md"
```

```python
with open(read_me, "r", encoding="utf-8") as f:
    readme_content = f.readlines()
```

```python
table_lines = [line.strip() for line in readme_content if "|" in line]
```

```python
if len(table_lines) > 1:
    table_lines.pop(1)

# Print the extracted lines
print("\n".join(table_lines))
```

```python
import re

records = []

for line in table_lines[1:]:
    try:
        upload_date = re.search(r"^\|\s*([0-9]{2}/[0-9]{2}/[0-9]{2})", line).group(1)

        # Extract title from markdown link
        title_match = re.search(r"\[(.*?)\]\(https:\/\/www\.youtube\.com\/watch\?v=.*?\)", line)
        title = title_match.group(1) if title_match else "UNKNOWN TITLE"

        # Extract transcript file name from transcript link
        file_match = re.search(r"(\d+__transcript\.txt)", line)
        file_name = file_match.group(1) if file_match else None

        if upload_date and title and file_name:
            records.append({
                "file_name": file_name,
                "upload_date": upload_date,
                "title": title
            })
    except Exception as e:
        print(f"Skipped line due to error: {e}\n{line}")

df_readme = pd.DataFrame(records)
print(df_readme.head())
```

```python
documents = []
for _, row in df_merged.iterrows():
    doc = {
        "title": row['title'],
        "upload_date": row['upload_date'],
        "file_name": row['file_name'],
        "content": row['text']
    }
    documents.append(doc)
```

```python
import nltk
from nltk.tokenize import TextTilingTokenizer
nltk.download('punkt')
nltk.download('stopwords')

#def semantic_split_with_texttiling(text, w=20):
    # TextTiling parameters: w is the size of the sliding window
#    tt = TextTilingTokenizer(w=w)

#    try:
#        chunks = tt.tokenize(text)
#        return chunks
#    except Exception as e:
#        fallback_splitter = RecursiveCharacterTextSplitter(
#            chunk_size=1000,
#            chunk_overlap=200
#        )
#        return fallback_splitter.split_text(text)




def splitter(text):
  splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200
        )
  return splitter.split_text(text)

chunked_documents = []
for doc in documents:
    # Split using TextTiling for topic-based chunks
    chunks = splitter(doc["content"])

    for i, chunk in enumerate(chunks):
        if len(chunk.split()) < 20:
            continue

        chunked_doc = {
            "title": doc["title"],
            "upload_date": doc["upload_date"],
            "file_name": doc["file_name"],
            "chunk_id": i,
            "content": chunk
        }
        chunked_documents.append(chunked_doc)
```

```python
import faiss
import pickle
from sentence_transformers import SentenceTransformer
from rank_bm25 import BM25Okapi
import json
import os
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

documents = chunked_documents
document_texts = [doc["content"] for doc in documents]
document_embeddings = model.encode(document_texts)


dimension = document_embeddings.shape[1]  # Should be 384 for all-MiniLM-L6-v2
index = faiss.IndexFlatIP(dimension) # Cosine distance
index.add(document_embeddings.astype('float32')) # Creating faiss index

# BM25
tokenized_corpus = [doc["content"].lower().split() for doc in documents]
bm25 = BM25Okapi(tokenized_corpus)
```

```python
os.makedirs("podcast_index", exist_ok=True)

# Save FAISS index
faiss.write_index(index, "podcast_index/faiss_index.bin")


with open("podcast_index/documents.pkl", "wb") as f:
    pickle.dump(documents, f)

with open("podcast_index/bm25.pkl", "wb") as f:
    pickle.dump(bm25, f)
```
